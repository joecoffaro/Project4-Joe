```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
file_path = Path("Resources/nsclc_tcga_broad_2016_clinical_data.tsv")
lung_cancer_df = pd.read_csv(file_path, sep="\t")
lung_cancer_df.head()
```

```python
lung_cancer_df["Sex"].head()
```

```python
lung_cancer_filled = lung_cancer_df.fillna(0)
lung_cancer_filled.head()
```

```python
lung_cancer_dropped_df = lung_cancer_filled.drop(["Study ID",
                                                 "Cancer Type",
                                                 "Number of Samples Per Patient",
                                                  "Somatic Status"
                                                 ],axis=1)
lung_cancer_dropped_df.head(50)
```

```python
lung_cancer_dropped_df['Cancer Type Detailed'] = lung_cancer_dropped_df['Cancer Type Detailed'].replace(
    {'Lung Adenocarcinoma': 1,
     'Lung Squamous Cell Carcinoma': 0})

lung_cancer_dropped_df['Prior Cancer Diagnosis Occurence'] = lung_cancer_dropped_df['Prior Cancer Diagnosis Occurence'].replace(
    {'No': 0,
     'Yes, history of prior malignancy': 1,
     'Yes, history of synchronous and or bilateral malignancy': 1})

lung_cancer_dropped_df['Sex'] = lung_cancer_dropped_df['Sex'].replace(
    {'Male': 0,
     'Female': 1
     })

lung_cancer_dropped_df['Smoking History'] = lung_cancer_dropped_df['Smoking History'].replace(
    {'Lifelong Non-Smoker': 0,
     'Current Smoker': 1,
     'Current Reformed Smoker For < Or = 15 Years': 1,
     'Current Reformed Smoker For > 15 Years':1,
     'Current Reformed Smoker, Duration Not Specified':1})
lung_cancer_dropped_df["Diagnosis Age"] = lung_cancer_dropped_df["Diagnosis Age"].astype("int64")
```

```python
dummies_df = pd.get_dummies(lung_cancer_dropped_df)
dummies_df
```

```python
clean_df = lung_cancer_dropped_df[['Diagnosis Age',
                                    'Fraction Genome Altered',
                                    'Mutation Count',
                                    'Overall Survival Status',
                                    'TMB (nonsynonymous)',
                                    'Cancer Type Detailed',
                                    'Prior Cancer Diagnosis Occurence',
                                    'Smoking History',
                                    'Sex',
                                    'Person Cigarette Smoking History Pack Year Value']]
clean_df.dropna()
clean_df
```

```python
clean_df.to_csv("Resources/cleaned_cancer_data.csv")
```
